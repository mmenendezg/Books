# Setup

```python
import matplotlib.pyplot as plt
import numpy as np

import tensorflow as tf
```

```python
IMG_HEIGHT = 224
IMG_WIDTH = 224
```

# A Dataset for Machine Perception

## Reading Image Data

```python
def read_and_decode(filename, reshape_dims):
    img = tf.io.read_file(filename) # Read the file
    img = tf.image.decode_jpeg(img, channels=3) # convert the compressed string to a 3d uint8 tensor
    img = tf.image.convert_image_dtype(img, tf.float32)
    return tf.image.resize(img, reshape_dims)
```

## Visualizing Image Data

```python
def show_image(filename):
    img = read_and_decode(filename, [IMG_HEIGHT, IMG_WIDTH])
    plt.imshow(img.numpy())
```

```python
show_image(
    'gs://cloud-ml-data/img/flower_photos/daisy/754296579_30a9ae018c_n.jpg'
)
```
