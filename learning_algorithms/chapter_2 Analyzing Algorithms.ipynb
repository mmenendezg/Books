## Using Empirical Models to Predict Performance

```python
def linear_model(n, a, b):
  return a*n + b

def quadratic_model(n, a, b):
  return a*n*n + b*n
```

```python
# Sample data
xs = [100, 1000, 10000]
ys = [0.063, 0.565, 5.946]

# Coefficients are returned as first argument
[(a,b), _] = curve_fit(linear_model, np.array(xs), np.array(ys))
print(f'Linear = {a}*N + {b}')
```

# Binary Array Search

```python
def binary_array_search(A, target):
  lo = 0
  hi = len(A) - 1

  while lo <= hi:
    mid = (lo + hi) // 2
    print(f'The index is {mid}, the target is {target} and the current value is {A[mid]}')

    if target < A[mid]:
      hi = mid - 1
    elif target > A[mid]:
      lo = mid + 1
    else:
      return True
  
  return False
```

```python
vector = [a for a in range(0, 10000000, 13)]
binary_array_search(vector, 17893)
```
